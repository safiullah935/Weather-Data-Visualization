```python
import numpy as np
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
temperature = [10, 12, 18, 24, 30, 35, 38, 37, 32, 25, 18, 12]
precipitation = [50, 40, 20, 10, 5, 2, 50, 80, 70, 30, 40, 60]

df = pd.DataFrame({'Month': months, 'Temperature': temperature, 'Precipitation': precipitation})

# Identify extreme values
max_temp = df['Temperature'].max()
min_temp = df['Temperature'].min()
max_rain = df['Precipitation'].max()

max_temp_month = df[df['Temperature'] == max_temp]['Month'].values[0]
min_temp_month = df[df['Temperature'] == min_temp]['Month'].values[0]
max_rain_month = df[df['Precipitation'] == max_rain]['Month'].values[0]

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# --- Subplot 1: Temperature Trend ---
axes[0].plot(df['Month'], df['Temperature'], marker='o', color='r', linestyle='-', linewidth=2, markersize=6, label='Temperature (C)')
axes[0].set_title("Monthly Temperature Trend", fontsize=14)
axes[0].set_xlabel("Month", fontsize=12)
axes[0].set_ylabel("Temperature (C)", fontsize=12)
axes[0].legend()
axes[0].grid(True)

# Annotate hottest and coldest months
axes[0].annotate(f"Hottest: {max_temp}C", 
                 xy=(max_temp_month, max_temp), 
                 xytext=(max_temp_month, max_temp + 2),
                 arrowprops=dict(facecolor='red', shrink=0.05), 
                 fontsize=10, color='red')

axes[0].annotate(f"Coldest: {min_temp}C", 
                 xy=(min_temp_month, min_temp), 
                 xytext=(min_temp_month, min_temp - 3),
                 arrowprops=dict(facecolor='blue', shrink=0.05), 
                 fontsize=10, color='blue')

# --- Subplot 2: Precipitation ---
axes[1].bar(df['Month'], df['Precipitation'], color='blue', alpha=0.6, label='Precipitation (mm)')
axes[1].set_title("Monthly Precipitation", fontsize=14)
axes[1].set_xlabel("Month", fontsize=12)
axes[1].set_ylabel("Precipitation (mm)", fontsize=12)
axes[1].legend()
axes[1].grid(axis='y', linestyle='--', alpha=0.7)

# Annotate rainiest month
axes[1].annotate(f"Rainiest: {max_rain}mm", 
                 xy=(max_rain_month, max_rain), 
                 xytext=(max_rain_month, max_rain + 10),
                 arrowprops=dict(facecolor='black', shrink=0.05), 
                 fontsize=10, color='black')

# Correct layout function
plt.tight_layout()

# Showplot
plt.show()
```
